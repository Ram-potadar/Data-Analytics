## Ploting Multiple Lines

## Chart Title and Legend

To differntiate between multiple lines it can be done using `plt.legend` function . Title can be given by the `plt.title()` function.

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

## Line markers

Markers can be shown by `marker` argument in the `plt.plot()` function.

```python
plt.plot(years, apples ,marker = 'o')
plt.plot(years, oranges , marker = 'x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples','oranges' ])
```

## Styling Lines and Markers

The `plt.plot` function supports many arguments for stypling line and markers:
- `color` or `c`: Set the color of the line ([supported colors](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` or `ls`: Choose between a solid or dashed line
- `linewidth` or `lw`: Set the width of a line
- `markersize` or `ms`: Set the size of markers
- `markeredgecolor` or `mec`: Set the edge color for markers
- `markeredgewidth` or `mew`: Set the edge width for markers
- `markerfacecolor` or `mfc`: Set the fill color for markers
- `alpha`: Opacity of the plot

```python
plt.plot(years, apples, marker ='s', c = 'b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker = 'o', c='r', ls='--', lw=2, ms=8, mew=2, mec='navy')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples','oranges' ])
```

`fmt` argument can be passed to the plt.plot function to specify the marker, line and color

### Bar Chart

Bar charts are similar to the line charts, i.e., they show a swquence of values. However, a bar is shown for each value, rather than points connected by the lines. We can use the `plt.bar` function to draw a bar chart.

```python
years  = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8 , 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
plt.plot(years, oranges);
```

This is line `chart`

```python
plt.bar(years, oranges);
```

We draw the both line chart and the bar chart together.

```python
plt.bar(years, oranges);
plt.plot(years, oranges, 'o--r');

plt.title('Yield of oranges');
```

Like histogram, bars can also be stacjed on the top of one another. We use the `bottom` argument to `plt.bar` to achive this.

```python
plt.bar(years, oranges);
plt.bar(years, oranges, bottom = oranges);
plt.legend(['Apples', 'Oranges']);
plt.title('Yeild of Apples and Oranges');
```

```python
plt.bar(years, oranges);
plt.bar(years, oranges, bottom = apples);
plt.legend(['Apples', 'Oranges']);
plt.title('Yeild of Apples and Oranges');
```
